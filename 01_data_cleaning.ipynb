```python
import pandas as pd
from pathlib import Path
```

```python
# Load in the raw pitch data.

raw_path = Path("../Data/Raw/Logan_Webb_Pitches.csv")

df = pd.read_csv(raw_path)

pd.set_option('display.max_columns', None)

df.head()
```

```python
# Sort every pitch in chronological order.

df = df.sort_values(
    by=[
        "game_date",
        "inning",
        "at_bat_number",
        "pitch_number"
    ]
).reset_index(drop=True)

df.head()
```

```python
# Keep only the columns needed for the analysis.

keep_columns = [
    "game_date",
    "inning",
    "at_bat_number",
    "pitch_number",
    "pitch_type",
    "pitch_name",
    "stand",
    "balls",
    "strikes",
    "release_speed",
    "effective_speed",
    "release_spin_rate",
    "spin_axis",
    "release_extension",
    "release_pos_x",
    "release_pos_z",
    "pfx_x",
    "pfx_z",
    "arm_angle",
    "plate_x",
    "plate_z",
    "sz_top",
    "sz_bot",
    "zone",
    "events",
    "description",
    "type",
    "launch_speed",
    "launch_angle",
    "launch_speed_angle",
    "hit_distance_sc",
    "estimated_ba_using_speedangle",
    "estimated_slg_using_speedangle",
    "estimated_woba_using_speedangle",
    "woba_value",
    "woba_denom",
    "babip_value",
    "iso_value",
    "hc_x",
    "hc_y"
]

df = df[keep_columns].copy()

df.head()
```

```python
# Label each pitch based on my manually selected good and bad starts.
# The good sample consists of Webb's five consecutive dominant starts in June.

df["start_group"] = "Excluded"

bad_starts = [
    "2026-03-25",
    "2026-04-11",
    "2026-04-17",
    "2026-05-05",
    "2026-07-03"
]

good_starts = [
    "2026-06-03",
    "2026-06-08",
    "2026-06-14",
    "2026-06-21",
    "2026-06-27"
]

df.loc[df["game_date"].astype(str).isin(bad_starts), "start_group"] = "Bad"
df.loc[df["game_date"].astype(str).isin(good_starts), "start_group"] = "Good"
```

```python
# Remove every pitch that is not part of the analysis (neither good nor bad).

df = df[df["start_group"] != "Excluded"].reset_index(drop=True)
```

```python
# Check for duplicate pitches.

duplicate_pitches = df.duplicated(
    subset=[
        "game_date",
        "inning",
        "at_bat_number",
        "pitch_number"
    ]
).sum()

duplicate_pitches
```

```python
# Check how many missing values are in each column.

df.isna().sum().sort_values(ascending=False)
```

```python
# Split the cleaned data into good and bad starts.

good_df = df[df["start_group"] == "Good"].copy()
bad_df = df[df["start_group"] == "Bad"].copy()
```

```python
# Check the size of each dataframe.

good_df.shape, bad_df.shape
```

```python
# Save the cleaned datasets.

good_df.to_csv("../Data/Cleaned/good_starts.csv", index=False)
bad_df.to_csv("../Data/Cleaned/bad_starts.csv", index=False)
```

```python
# Save the cleaned master dataset.

df.to_csv("../Data/Cleaned/all_selected_starts.csv", index=False)
```
